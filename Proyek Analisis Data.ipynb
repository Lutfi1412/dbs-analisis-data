### Assessing Data

```python
dataframes = {
    "Customers": customers_df,
    "Geolocation": geolocation_df,
    "Order Items": order_items_df,
    "Order Payments": order_payments_df,
    "Order Reviews": order_reviews_df,
    "Orders": orders_df,
    "Product Category": product_category_df,
    "Products": products_df,
    "Sellers": sellers_df,
}

for name, path in dataframes.items():
    duplicate_count = path.duplicated().sum()

    missing_values = path.isnull().sum()

    outlier_count = 0
    for col in path.select_dtypes(include=np.number):
        data = path[col].dropna()
        if len(data) > 0:
            q25, q75 = np.percentile(data, 25), np.percentile(data, 75)
            iqr = q75 - q25
            cut_off = iqr * 1.5
            minimum, maximum = q25 - cut_off, q75 + cut_off
            outliers = [x for x in data if x < minimum or x > maximum]
            outlier_count += len(outliers)

    print(f"Dataset: {name}")
    print(f"Jumlah Data Duplikat: {duplicate_count}")
    print(f"Jumlah Data Kosong (Missing Values): \n{missing_values}")
    print(f"Jumlah Outlier: {outlier_count}")
    print()
```

**Insight:**
**Data Duplikat**
         **Dataset yang memiliki duplikasi:**
         **Geolocation Dataset** → Terdapat beberapa baris yang memiliki nilai yang sama.
        
**Data Kosong (Missing Values)**
         **Kolom yang memiliki nilai kosong dalam beberapa dataset:**
         **Order Reviews Dataset**  
                 `review_comment_title`, `review_comment_message`  
         **Orders Dataset**  
                 `order_approved_at`, `order_delivered_carrier_date`, `order_delivered_customer_date`  
         **Products Dataset**  
                 `product_category_name`, `product_name_length`, `product_description_length`,  
                 `product_photos_qty`, `product_weight_g`, `product_length_cm`,  
                 `product_height_cm`, `product_width_cm`
        
**Data Outlier**
        `Geolocation Dataset`,
         `Products Dataset`,
         `Order Reviews Dataset`,
         `Order Items Dataset`,
         `Order Payments Dataset`

### Cleaning Data

```python
cleaned_dataframes = {}
for name, df in dataframes.items():
    df = df.copy()

    df.drop_duplicates(inplace=True)

    for col in df.columns:
        if df[col].isnull().sum() > 0:
            if df[col].dtype == "object":
                df.dropna(subset=[col], inplace=True)
            else:
                df[col] = df[col].fillna(df[col].mean())

    for col in df.select_dtypes(include=np.number):
        data = df[col].dropna()
        if len(data) > 0:
            q25, q75 = np.percentile(data, 25), np.percentile(data, 75)
            iqr = q75 - q25
            cut_off = iqr * 1.5
            minimum, maximum = q25 - cut_off, q75 + cut_off
            df[col] = df[col].mask((df[col] < minimum) | (df[col] > maximum), maximum)

    cleaned_dataframes[name] = df
    print(f"Dataset {name} telah dibersihkan.")
```

**Insight:**

data data yang tadi nya duplicate, outliner, dan null kemudian di cleaning.

**Data duplikat** telah dihapus.

**Data kosong** telah ditangani: jika tipe data *object*, maka baris tersebut dihapus; jika numerik, nilai kosong diganti dengan rata-rata.

**Outlier** telah ditangani dengan menggantinya menggunakan nilai maksimum dalam rentang IQR.

## Exploratory Data Analysis (EDA)

### Explore ...

```python
all_df = pd.merge(
    left=cleaned_dataframes["Customers"],
    right=cleaned_dataframes["Orders"],
    how="inner",
    left_on="customer_id",
    right_on="customer_id"
)

rating_state = pd.merge(
    left=all_df,
    right=cleaned_dataframes["Order Reviews"],
    how="inner",
    left_on="order_id",
    right_on="order_id"
)

state_review_analysis = rating_state.groupby("customer_state")["review_score"].mean().reset_index()

state_review_analysis = state_review_analysis.sort_values(by="review_score", ascending=False)

display(state_review_analysis)
```

```python
products_state = pd.merge(
    left=rating_state,
    right=cleaned_dataframes["Order Items"],
    how="inner",
    left_on="order_id",
    right_on="order_id"
)

products_state = pd.merge(
    left=products_state,
    right=cleaned_dataframes["Products"],
    how="inner",
    left_on="product_id",
    right_on="product_id"
)

state_products_analysis = products_state.groupby(["customer_state", "product_category_name"]).size().reset_index(name="total_purchases")

top_product_per_state = state_products_analysis.loc[state_products_analysis.groupby("customer_state")["total_purchases"].idxmax()]

display(top_product_per_state)

```

**Insight:**

**Rata-rata Rating per Negara:**  
untuk explore data pertama saya menggunakan data customer dan order yang sudah di cleaning kemudian 2 data tersebut di cocokan berdasarkan kolom custumer_id nya, kemudian 2 data yang telah di satukan di cocokan kembali dengan data order review, lalu saya cocok kan kembali data nya dengan kolom order_id.

Dari semua data yang telah di satukan lalu data nya saya gruping untuk kolom custumer_state dan mengambil nilai rata rata dari kolom review_score setiap gruping state tersebut.

 **Analisis Produk Terlaris setiap Negara**  
untuk explore data kedua saya menggunakan data yang telah di satukan sebelumnya dan data Order Items saya mengambil value yang sama berdasarkan kolom order_id, kemudian 2 data yang telah di satukan di cocokan kembali dengan data products, lalu saya cocok kan kembali data nya dengan kolom product_id.

Dari semua data yang telah di satukan lalu data nya saya gruping untuk kolom custumer_state dan product_category_name kemudian saya hitung setiap index nya dan membuat kolom total_purchases untuk laporan produk terlarisnya, setelah itu ambil nilai terbanyak productnya





## Visualization & Explanatory Analysis

### Pertanyaan 1:
Berapa rata rata rating yang di berikan pada setiap customers pada suatu negara?


```python
plt.figure(figsize=(12, 6))
plt.bar(
    state_review_analysis["customer_state"],
    state_review_analysis["review_score"],
    color='skyblue'
)

plt.xlabel("Negara")
plt.ylabel("Rata-rata Review Score")
plt.title("Rata-rata Review Score di Setiap Negara")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

### Pertanyaan 2:
Barang apa saja yang laku terjual pada setiap negara?

```python
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=top_product_per_state,
    x="customer_state",
    y="total_purchases",
    hue="product_category_name",
    marker="o",
    linewidth=2
)

plt.xticks(rotation=45)
plt.xlabel("Customer State")
plt.ylabel("Total Purchases")
plt.title("Produk Paling Banyak Dibeli di Tiap Negara (Line Chart)")
plt.legend(title="Product Category", bbox_to_anchor=(1, 1))
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

untuk barang yang laris terjual untuk setiap negara saya juga memvisualisasikannya, namun untuk data ini saya menggunakan metode line chart untuk menyajikan datanya.
